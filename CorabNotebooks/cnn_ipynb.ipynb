```python
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPool2D
from keras.optimizers import Adam

from keras.layers.core import Dense, Activation, Dropout, Flatten
from keras.utils import plot_model
from keras.callbacks import TensorBoard

from keras.datasets import cifar10
from keras.utils import np_utils
```

```python
(X_train, y_train),(X_test, y_test) = cifar10.load_data()
```

```python
# floatに変換しておく
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# 正規化
X_train /= 255.0
X_test /= 255.0

# one-hot vector に変換したほうが都合が良い
n_classes = 10
Y_train = np_utils.to_categorical(y_train, n_classes)
Y_test = np_utils.to_categorical(y_test, n_classes)
```

```python
# モデルの定義
model = Sequential()

model.add(Conv2D(32,(5,5), padding='same', input_shape=(32,32,3)))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2),padding = 'same'))
model.add(Conv2D(16,(10,10), padding='same'))
model.add(Activation('relu'))
model.add(MaxPool2D(pool_size=(2,2),padding = 'same'))

model.add(Flatten())
model.add(Dense(1024))
model.add(Activation('relu'))

model.add(Dense(n_classes, activation='softmax'))

adam = Adam(lr=1e-3)

model.compile(optimizer=adam, loss='categorical_crossentropy', metrics=["accuracy"])

model.summary()
```

```python
batch_size = 50
nb_epoch = 10
validation_split=0.1

history = model.fit(X_train, Y_train, batch_size=batch_size, epochs=nb_epoch, verbose=1, validation_split=validation_split)
```
